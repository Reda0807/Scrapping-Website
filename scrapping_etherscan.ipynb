```python
# Install chromium, its driver, and selenium
# !apt-get update
# !apt install chromium-chromedriver
# !pip install selenium

# Set up the path to the chromedriver executable
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import random
import string

# Set the page limit
strr = 'ffffffff'
str1 = strr + strr + strr + strr + strr + strr + strr + strr
limit_page = int(int(str1, 16)) // 60

def configure_chrome_options():
    # Configure options to run Chrome in headless mode
    options = Options()
    options.add_argument('--headless')        # Runs Chrome in headless mode.
    options.add_argument('--no-sandbox')      # Bypass OS security model
    options.add_argument('--disable-dev-shm-usage')  # Overcome limited resource problems
    return options

# Instantiate a webdriver
options = configure_chrome_options()
fetch_driver = webdriver.Chrome(options=options)

def get_random_page():
    # Generate a random 4-character string with 0-9 and a-f
    rd_str1 = ''.join(random.choices(string.hexdigits[:16], k=16))
    rd_str2 = ''.join(random.choices(string.hexdigits[:16], k=16))
    rd_str3 = ''.join(random.choices(string.hexdigits[:16], k=16))
    rd_str4 = ''.join(random.choices(string.hexdigits[:16], k=16))

    random_str = rd_str1 + rd_str2 + rd_str3 + rd_str4

    # Convert the string to a hexadecimal number
    hex_num = int(random_str, 16)

    # Convert the hexadecimal number to a decimal number
    decimal_num = int(hex_num) // 60

    return decimal_num

def fetch_page_title(driver, url):
    # Navigate to the page
    driver.get(url)

    try:
      wait = WebDriverWait(driver, 30)
      # Assuming 'badge' class uniquely identifies the span with "0 ether"
      # span_element = driver.find_element('css selector', 'span#total-balance')
      span_element = wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, 'span#total-balance')))
      text = span_element.text
      return text
    except Exception as e:
      print(f"An error occurred: {e}")

def random_privateKey():
    # Generate a random 4-character string with 0-9 and a-f
    rd_str1 = ''.join(random.choices(string.hexdigits[:16], k=16))
    rd_str2 = ''.join(random.choices(string.hexdigits[:16], k=16))
    rd_str3 = ''.join(random.choices(string.hexdigits[:16], k=16))
    rd_str4 = ''.join(random.choices(string.hexdigits[:16], k=16))

    random_str = rd_str1 + rd_str2 + rd_str3 + rd_str4
    # Convert the string to a hexadecimal number
    hex_num = int(random_str, 16)

    # Convert the hexadecimal number to a decimal number
    decimal_num = int(hex_num)

    return decimal_num


# Usage example
base_url = "https://privatekeyfinder.io/private-keys/ethereum/"
count = 5000000
for i in range(count):
  page_num = get_random_page()
  if page_num > limit_page:
    continue
  url = base_url + str(page_num)
  content = fetch_page_title(fetch_driver, url)
  if content.endswith('ether'):
    ether_value = float(content.split()[0])
#    if ether_value > 0.01:
  print(f"The url of this page {url} is: {content}")

fetch_driver.quit()
```
